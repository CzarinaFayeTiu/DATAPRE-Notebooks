```python
import pprint
import requests
import datetime
import pandas as pd

#apikey 
secret = "733dd542bffb474082e59bb4ea948919"
   
# Define the endpoint
url = 'https://newsapi.org/v2/everything?'
   
#define query, api, to and from datetime
parameters = {
    'q': 'Philippines', # query phrase
    'apiKey': secret, # your own API key
    'from': '2022-5-21T00:00:00Z',
    'to': '2022-5-22T23:59:59Z'
}
   
# Make the request
response = requests.get(url,  params = parameters)

#convert request response in json
response_json = response.json()

#array to store all responses 
extracted_api_info = []

#load responses (title, author, published date, and content into array)
for i in response_json['articles']:
    title = i['title']
    author = i['author']
    publishedAt = i['publishedAt']
    fullArticle = i['content']
    extracted_api_info.append((title, author, publishedAt, fullArticle))
    
```

```python
df_api_list = pd.DataFrame(extracted_api_info)
df_api_list.columns = ['Title', 'Author', 'PublishedAt', 'FullArticle']
df_api_list.to_json('TIU_MAGTIBAY_api.json',orient='index')
pd.read_json('TIU_MAGTIBAY_api.json')
```
